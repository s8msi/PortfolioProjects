### 2. Converting the data type of 'gwTz' to datetime.

### 4. Trying to find out when the population sleeps.

```python
plt.figure(figsize = [7,6])
data.rcvdTm.dt.hour.value_counts().plot.barh(color = 'darkgreen')
plt.show()
```

```python
plt.figure(figsize = [10,5])
data.sloc[data.rcvdTm.dt.hour == 21].value_counts().plot.line(color = 'red')
plt.show()
```

Since there are 24 hours in a day, let's break them down into 4 parts to analyze better. 0-5, 6-11, 12-17 & 18-23.

```python
data.sloc[data.rcvdTm.dt.hour <= 5].value_counts().plot.bar()
plt.show()
```

```python
data.sloc[(data.rcvdTm.dt.hour >= 6) & (data.rcvdTm.dt.hour < 12)].value_counts().plot.bar()
plt.show()
```

```python
data.sloc[(data.rcvdTm.dt.hour >= 12) & (data.rcvdTm.dt.hour < 18)].value_counts().plot.bar()
plt.show()
```

```python
data.sloc[data.rcvdTm.dt.hour >= 18].value_counts().plot.bar()
plt.show()
```

After a bit of trial and error, I found out the time they sleep. They prefer to sleep between 22:00 - 05:00 (or they are at least in the bedroom) because the graph between this time is indicating only towards the bedroom.

```python
data.sloc[data.rcvdTm.dt.hour >= 22].value_counts().plot.bar()
plt.show()
```

```python
data.sloc[data.rcvdTm.dt.hour <= 5].value_counts().plot.bar()
plt.show()
```

```python
((data.sloc[data.rcvdTm.dt.hour <= 5].value_counts().max() + data.sloc[data.rcvdTm.dt.hour >= 22].value_counts().max())/(data.sloc[data.rcvdTm.dt.hour <= 5].value_counts().sum() + data.sloc[data.rcvdTm.dt.hour >= 22].value_counts().sum())) * 100
```

Since we can see from the visualizations above that the bar-graph has most of the movements recorded in the bedroom, I tried to find the percentage of the population that is asleep. I found out that 92% of the population are in the bedroom in between 22:00 - 06:00. Since 92% is a huge portion of the population, we can finalize that the senior citizens in the UK, aged 80 and above sleep in between 22:00 - 06:00!

### 5. Let's decide how to find out the area 'Out of Location'.

```python
data.rcvdTm.dt.hour[(data.sloc == 'Out of Location')].value_counts().sort_values().plot.bar()
plt.show()
```

We can clearly see here that the sensors have been triggered in the area 'Out of Location' only once between 08:00 - 09:00 & 17:00 - 18:00 each. Which makes a total of only 2 out of 37 times that the sensor was triggered. So, this can be ignored as it won't affect the analysis. Whereas, for the rest of the time, i.e... between 09:00 - 17:00, the motion sensor has been triggered for a total of 35 times. Therefore, we need to set up a camera in their homes which is active only between 09:00 - 17:00. After that when the sensor triggers in the area, 'Out of Location', we can verify that with the time on the camera to find out the area that is 'Out of Location'.

Therefore, these are my two findings:
1. They sleep between 22:00 - 06:00.
2. We need to set up a camera between 09:00 - 17:00 to find out the location 'Out of Location'.
