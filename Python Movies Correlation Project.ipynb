```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create.

# Reading in the data

df = pd.read_csv(r'C:\Users\ALPHONSA\Downloads\movies.csv')
```

```python
# Checking if any columns have null values
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Checking the data types of the columns
```

```python
df.dtypes
```

```python
# Changing the data type of the columns: budget and gross.
```

```python
# df['budget'] = df['budget'].astype('int64')
# df['gross'] = df['gross'].astype('int64')

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 

df.head()
```

```python
# Ordering the data by Gross.

df.sort_values(by = ['gross'], inplace = False, ascending = False)
```

```python
# Since there are 3 faulty rows at the end, we remove them.

df.drop(df[df['gross'] < 0].index, inplace = True)
df.sort_values(by = ['gross'], inplace = False, ascending = False)

# -> To display all the rows || pd.set_option('display.max_rows', None)
```

```python
# Checking for duplicates
df['name'].drop_duplicates().sort_values(ascending = False)

# To remove duplicates -> df.drop_duplicates()
```

```python
# Scatter plot of budget and gross.

plt.scatter(x = df['budget'], y = df['gross'], color = 'red')
plt.title('Budget v Gross earnings')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.show()
```

```python
# Budget and gross on a regression plot.

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "black"}, line_kws = {"color": "blue"})
```

```python
# The correlation.

df.corr() # pearson, kendall, spearman
```

```python
# Correlation matrix on a Heatmap.

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlatoin matrix of movies')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
# Converting all the data to numbers.

df_num = df

for col_name in df_num.columns:
    if(df_num[col_name].dtype == 'object'):
        df_num[col_name] = df_num[col_name].astype('category')
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num.sort_values(by =['gross'], ascending = False)
```

```python
# Correlaton matrix of all the data in the table.

correlation_matrix1 = df_num.corr()
sns.heatmap(correlation_matrix1, annot = True)
plt.title('Correlatoin matrix of movies')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
plt.show()
```

```python
correlation_matrix1
```

```python
# Correlation matrix in pairs

corr_pairs = correlation_matrix1.unstack()
pd.set_option('display.max_rows', None)
corr_pairs
```

```python
# Sorted pairs

sorted_pairs = corr_pairs.sort_values()
sorted_pair = sorted_pairs[sorted_pairs > 0.5]
sorted_pair[sorted_pair < 1]
```
